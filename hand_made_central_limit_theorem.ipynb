```python
import numpy
import scipy.stats as sts
import pandas
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Используем Бета распределение 
# Задаем параметры распределения и генерируем выборку объема 1000
a = 2
b = 5
beta_distr = sts.beta(a, b)
sample = beta_distr.rvs(1000)

# Генерируем теоретическую плотность распределения случайной величины(theoretical pdf)
space = numpy.linspace(0, 1, 100)
pdf = beta_distr.pdf(space)

# Строим гистограмму выборки и теоретическую плотность распределения случайной величины
plt.hist(sample, density=True, bins=40, label='beta destribution')
plt.plot(space, pdf, label='theoretical pdf')
plt.legend()
plt.ylabel('fraction of samples, $F(x)$')
plt.xlabel('$x$')
```

```python
# Генерируем 1000 выборок объемов 5, 10, 30, 50 и для каждой находим их выборочные средние
mid = {'5': [], '10': [], '30': [], '50': []}
for each in ['5', '10', '30', '50']:
    for i in range(1000):
        mid[each].append(sum(beta_distr.rvs(int(each)))/int(each)) 
# находим теоретические мат.ожидание и дисперсию бета распределения
m = a/(a + b)
S_theoretical = (a * b)/(((a + b)**2)*(a + b + 1))

# вычисляем параметры нормальных распределений(перерасчет дисперсии), аппроксимирующих выборочные средние при различных n(5, 10, 30, 50)
# генерируем нормальные распределения и находим плотности этих распределений
pdf = []
for each in [5, 10, 30, 50]:
    S = numpy.sqrt(S_theoretical/each)
    norm_d = sts.norm(m, S)
    pdf.append(norm_d.pdf(space))
```

```python
# гистограмма распределения выборочных средних для выборки объема 5 и плотность соответств. нормального распределения
plt.hist(mid['5'], density=True, bins=20, label="выборочные средние")
plt.plot(space, pdf[0], label="плотность норм. распред. n=5")
plt.legend()
plt.ylabel('F(x)')
plt.xlabel('$x$')
```

```python
# гистограмма распределения выборочных средних для выборки объема 10 и плотность соответств. нормального распределения
plt.hist(mid['10'], density=True, bins=20, label="выборочные средние")
plt.plot(space, pdf[1], label="плотность норм. распред. n=10")
plt.legend()
plt.ylabel('F(x)')
plt.xlabel('$x$')
```

```python
# гистограмма распределения выборочных средних для выборки объема 30 и плотность соответств. нормального распределения
plt.hist(mid['30'], density=True, bins=30, label="выборочные средние")
plt.plot(space, pdf[2], label="плотность норм. распред. n=30")
plt.legend()
plt.ylabel('F(x)')
plt.xlabel('$x$')
```

```python
# гистограмма распределения выборочных средних для выборки объема 50 и плотность соответств. нормального распределения
plt.hist(mid['50'], density=True, bins=30, label="выборочные средние")
plt.plot(space, pdf[3], label="плотность норм. распред. n=50")
plt.legend()
plt.ylabel('F(x)')
plt.xlabel('$x$')
```
